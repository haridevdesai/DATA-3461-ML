When analyzing time-series data such as the Bitcoin Weighted Price Over Time depicted in the plot, a critical step in the preparatory data analysis is assessing the stationarity of the series. Stationarity in a time series means that its statistical properties such as mean, variance, and autocorrelation are all constant over time. Non-stationary time series data can lead to misleading statistics and poor predictions.

For the Bitcoin data shown, the visible trends and fluctuations in price, especially the sharp spike post-2020, suggest that the series could be non-stationary. This non-stationarity is often characteristic of financial time series due to factors like market dynamics, economic changes, and investor behavior influencing the prices.

To quantitatively determine the stationarity of the Bitcoin price series, we can apply the Augmented Dickey-Fuller (ADF) test, a widely used statistical test. The null hypothesis of the ADF test is that the time series is non-stationary. Therefore, a higher p-value (>0.05) would indicate that we fail to reject the null hypothesis, confirming the non-stationarity of the data. Conversely, a low p-value (≤0.05) suggests rejecting the null hypothesis, indicating stationarity.

The implications of finding non-stationarity are significant for forecasting. It necessitates differencing the series, applying transformations, or using models like ARIMA, which incorporate differencing to handle such data. Ensuring that the time series is stationary is crucial for the reliability of the predictions and the overall effectiveness of the forecasting model. In practice, handling non-stationarity before applying predictive models is essential for achieving accurate and reliable forecasts in volatile markets like that of Bitcoin.

```python
#check with Dickey-Fuller test
#H0: The time series is non-stationary.
#H1: The time series is stationary.
#0.05 significance level
# Perform Augmented Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(df['Weighted_Price'].dropna())

print('ADF Statistic: %f' % adf_test[0])
print('p-value: %f' % adf_test[1])
print('Critical Values:')
for key, value in adf_test[4].items():
    print('\t%s: %.3f' % (key, value))

# Interpretation:
if adf_test[1] < 0.05:
    print("The time series is stationary.")
else:
    print("The time series is not stationary.")
```

Now since we got a higher p-value, let's do some transformations to make the series more stationary so that it is better for our model.

```python
# Perform differencing to make the data stationary
df['Weighted_Price_diff'] = df['Weighted_Price'].diff().dropna()

# Check if the differenced series is stationary
adf_test_diff = adfuller(df['Weighted_Price_diff'].dropna()) # Make sure to drop NaN values after differencing

print('ADF Statistic for Differenced Series: %f' % adf_test_diff[0])
print('p-value for Differenced Series: %f' % adf_test_diff[1])
print('Critical Values for Differenced Series:')
for key, value in adf_test_diff[4].items():
    print('\t%s: %.3f' % (key, value))
```

## ACF and PACF Graphs
The Autocorrelation Function (ACF) and the Partial Autocorrelation Function (PACF) are key tools in the identification of the order of Autoregressive (AR) and Moving Average (MA) processes in time series analysis.

ACF is a measure of the correlation between the time series (ts) with a lagged version of itself. For instance, ACF(k) will give the correlation of the series with its k-th lag. This can be used to identify the MA(q) component of an ARIMA model, as the ACF theoretically cuts off after the q-th lag for an MA(q) process.

PACF, on the other hand, measures the partial correlation of the series with its own lagged values, controlling for the values of the time series at all shorter lags. It essentially removes the effect of earlier lags. This can be useful in identifying the AR(p) component of an ARIMA model, as the PACF theoretically cuts off after the p-th lag for an AR(p) process.

```python
#acf graph
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df['Weighted_Price_diff'], lags=100)
plt.show()
```

# Train-Test Split

```python
#train test split
train = df.loc[df.index < '2020-01-01']
test1 = df.loc[df.index >= '2020-01-01']
test = df.loc[(df.index >= '2020-01-01') & (df.index < '2020-12-01')]

fig, ax = plt.subplots(figsize=(10, 4))
train['Weighted_Price'].plot(ax=ax, label='Training set')
test['Weighted_Price'].plot(ax=ax, label='Test set')
plt.show()

fig, ax = plt.subplots(figsize=(10, 4))
train['Weighted_Price'].plot(ax=ax, label='Training set')
test1['Weighted_Price'].plot(ax=ax, label='Test set')
plt.show()
```

For our predictive modeling, we chose to use the first test dataset, primarily because it displayed a more stable and consistent trend over time without significant spikes. The absence of such anomalies in the dataset allows us to evaluate the model’s performance under 'normal' conditions, where external factors are less likely to skew the results. This approach is beneficial for understanding how well our model can capture and predict regular patterns and behaviors inherent to the data.

The decision to avoid using the second dataset, characterized by a significant spike around 2020-2021, was strategic. This spike, likely influenced by external factors not accounted for in our models (e.g., economic changes, market disruptions), would likely lead to poor model performance. Models like SARIMA, which assume a certain level of continuity and seasonality in data, might find it challenging to predict such abrupt changes accurately. Similarly, even machine learning models, unless explicitly designed to factor in such anomalies, might fail to generalize well in the presence of these spikes.

This approach underscores a crucial aspect of predictive modeling: the selection of training and testing data can profoundly impact the model's perceived accuracy and generalizability. By choosing a dataset that represents typical conditions, we aim to develop a model that reliably predicts future data under similar circumstances. However, it also highlights a limitation — our current models might not yet be equipped to handle abrupt, significant changes driven by external shocks. Moving forward, enhancing our models to anticipate and respond to such anomalies will be an essential step in improving their robustness and utility in real-world applications.

```python
# Mean Absolute Error (MAE)
mae_xgb = mean_absolute_error(y_test, xgb_predictions)

# Mean Squared Error (MSE)
mse_xgb = mean_squared_error(y_test, xgb_predictions)

# Root Mean Squared Error (RMSE)
rmse_xgb = np.sqrt(mse_xgb)

# Mean Absolute Percentage Error (MAPE)
mape_xgb = np.mean(np.abs((y_test - xgb_predictions) / y_test)) * 100

# Symmetric Mean Absolute Percentage Error (sMAPE)
smape_xgb = 100 * np.mean(2.0 * np.abs(y_test - xgb_predictions) / (np.abs(y_test) + np.abs(xgb_predictions)))

# Mean Absolute Scaled Error (MASE)
# For MASE we need the naive forecast (usually the previous value in the series)
naive_forecast_xgb = np.roll(y_test, 1)
mae_naive_xgb = mean_absolute_error(y_test[1:], naive_forecast_xgb[1:])
mase_xgb = mae_xgb / mae_naive_xgb

# R-squared (R²)
r2_xgb = r2_score(y_test, xgb_predictions)

# Adjusted R-squared - requires the number of predictors (k) and sample size (n)
k = X_test.shape[1]  # number of predictors
n = len(y_test)
adjusted_r2_xgb = 1 - (1 - r2_xgb) * (n - 1) / (n - k - 1)

# Forecast Bias
bias_xgb = np.mean(xgb_predictions - y_test)

# Print the metrics
print(f"MAE: {mae_xgb}")
print(f"MSE: {mse_xgb}")
print(f"RMSE: {rmse_xgb}")
print(f"MAPE: {mape_xgb}")
print(f"sMAPE: {smape_xgb}")
print(f"MASE: {mase_xgb}")
print(f"R-squared: {r2_xgb}")
print(f"Adjusted R-squared: {adjusted_r2_xgb}")
print(f"Forecast Bias: {bias_xgb}")
```

## SARIMA


```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
from statsmodels.graphics.tsaplots import plot_predict
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error

# Perform the auto_arima function to find the best SARIMA parameters
stepwise_fit = auto_arima(df['Weighted_Price'], trace=True, suppress_warnings=True)
print(stepwise_fit.summary())
```

In our effort to optimize our time series forecasting, we employed the `auto_arima` function from the `pmdarima` library, which automates the process of identifying the most suitable ARIMA model by systematically testing different combinations of parameters. This function evaluates various seasonal and non-seasonal components of the ARIMA models to determine the best fit based on the Akaike Information Criterion (AIC), a well-known measure used to select models while balancing goodness of fit and model complexity.

The output of the `auto_arima` function revealed that the best model for our data was the ARIMA(0,2,1)(0,0,0)[0], which suggests a second order differencing with a first order moving average component and no seasonal effects. This model configuration, selected for having the lowest AIC, indicates that it provides a strong balance between fitting the data well and not overfitting by keeping the model relatively simple.

Despite the strong statistical backing from the ARIMA selection process, this model did not perform as well as anticipated in practical forecasting scenarios when compared to other models we tested like the Random Forest and XGBoost. The prediction errors, as measured by metrics such as the mean squared error, were higher than those produced by these machine learning models. This discrepancy might be due to the intrinsic properties of the data, such as non-linear relationships that SARIMA models, which assume linearity, fail to capture effectively.

Given these outcomes, it became clear that while ARIMA models are powerful for many time series datasets, they might not be the best choice for all types of data, particularly those where complex patterns and relationships exist that are better handled by more flexible, non-parametric models. Our experience thus underscores the importance of model testing and validation in real-world scenarios, beyond theoretical statistical metrics, to ensure optimal performance in predictive tasks.

Moreover, the Random Forest model showed a remarkable R² and Adjusted R², nearly perfect, which suggests that it can explain nearly all the variability in the response variable by the predictors. On the contrary, the XGBoost, while still performing well, fell short in these metrics, pointing to a potential overfit on the training data, or it may not generalize as effectively on unseen data.

The Forecast Bias also highlighted a stark contrast, with Random Forest displaying a small positive bias, whereas XGBoost showed a substantial negative bias, indicating a tendency to underestimate the target variable significantly.

In summary, these results highlight the Random Forest's superior performance in our specific context, suggesting it as the preferable model for predictions in this scenario. We should consider these findings when deciding which model to deploy, particularly in environments where precision is crucial. Further investigations into parameter tuning and feature engineering for the XGBoost model might help reduce its forecast bias and improve its overall predictive power.
